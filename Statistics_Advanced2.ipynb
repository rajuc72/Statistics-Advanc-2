####1. **What is hypothesis testing in statistics?**

Ans: Test an assumption or claim about a population parameter. It's a way to use sample data to assess the validity of a hypothesis, often about the mean or proportion of a population.

####2.**What is the null hypothesis, and how does it differ from the alternative hypothesis?**
Ans: The default assumption that there is no effect, no difference, or no relationship between variables is called a Null Hypothesis

####3**.What is the significance level in hypothesis testing, and why is it important?**
Ans: The significance level (denoted as alpha) in hypothesis testing is a critical threshold that determines the probability of rejecting the null hypothesis (H₀) when it is actually true. It's a key element of hypothesis testing.

####4.**What does a P-value represent in hypothesis testing?**
Ans: A p-value in hypothesis testing measures the strength of the evidence against the null hypothesis (H₀).

####5.**How do you interpret the P-value in hypothesis testing?**
Ans:The p-value is compared to the significance level. If p-value < alpha, you reject H₀; otherwise, you fail to reject H₀.

####6.**What are Type 1 and Type 2 errors in hypothesis testing?**
Ans: Type1 error occurs when you reject the null hypothesis (H₀) when it is actually true.
Type2 error occurs when you fail to reject the null hypothesis (H₀) when it is actually false.

A two-tailed test (or two-sided test) examines whether a parameter is different from a certain value, without specifying the direction of the difference.

###1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
import scipy.stats as stats

# Sample data
sample_mean = 165
population_mean = 170
population_std_dev = 10
sample_size = 25
# Calculate the standard error
standard_error = population_std_dev / np.sqrt(sample_size)
# Calculate the Z-score
z_score = (sample_mean - population_mean) / standard_error
# Calculate the p-value
p_value = stats.norm.sf(abs(z_score)) * 2  # Two-tailed test
# Set the significance level
alpha = 0.05
# Output the results
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

###3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
# Step 1: Simulate random data
np.random.seed(42)  # For reproducibility
population_mean = 50
population_std = 5
sample_size = 30
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
# Step 2: Calculate the sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Use ddof=1 for sample standard deviation
# Step 3: Perform the one-sample Z-test
# Null Hypothesis (H0): The sample mean is equal to the population mean (50)
# Alternative Hypothesis (H1): The sample mean is not equal to the population mean
# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
# Calculate the P-value
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))
# Step 4: Output the results
print(f"Sample Data: {sample_data}")
print(f"Mean of Sample Data: {sample_mean}")
print(f"Z-Statistic: {z_statistic}")
print(f"P-Value: {p_value}")
# Step 5: Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis (H0). The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0). The sample mean is not significantly different from the population mean.")
```

###4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

###5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_errors(population_mean, population_std, sample_size, true_mean, alpha=0.05):
    # Step 1: Calculate the critical value for the given alpha (two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Step 2: Simulate sample data based on the true mean
    np.random.seed(42)
    sample_data = np.random.normal(loc=true_mean, scale=population_std, size=sample_size)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)

    # Step 3: Calculate the Z-statistic for the sample mean
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Step 4: Calculate Type 1 error (alpha) and Type 2 error (beta)
    type_1_error = alpha
    type_2_error = 1 - stats.norm.cdf(z_critical - (true_mean - population_mean) / (population_std / np.sqrt(sample_size)))

    # Step 5: Visualize the distributions and errors
    x = np.linspace(-4, 4, 1000)
    y_null = stats.norm.pdf(x)
    y_alternative = stats.norm.pdf(x - (true_mean - population_mean) / (population_std / np.sqrt(sample_size)))

    plt.plot(x, y_null, label="Null Hypothesis (H0)")
    plt.plot(x, y_alternative, label="Alternative Hypothesis (H1)")
    plt.fill_between(x, 0, y_null, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5, label="Type 1 Error (alpha)")
    plt.fill_between(x, 0, y_alternative, where=(x >= -z_critical) & (x <= z_critical), color='blue', alpha=0.5, label="Type 2 Error (beta)")
    plt.axvline(z_statistic, color='green', linestyle='--', label=f"Z-statistic = {z_statistic:.2f}")
    plt.axvline(-z_critical, color='black', linestyle='--', label=f"Z-critical = {-z_critical:.2f}")
    plt.axvline(z_critical, color='black', linestyle='--', label=f"Z-critical = {z_critical:.2f}")
    plt.legend()
    plt.title("Type 1 and Type 2 Errors Visualization")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.show()

    # Output results
    print(f"Sample Data: {sample_data}")
    print(f"Mean of Sample Data: {sample_mean}")
    print(f"Z-Statistic: {z_statistic}")
    print(f"P-Value: {2 * (1 - stats.norm.cdf(np.abs(z_statistic)))}")
    print(f"Type 1 Error (alpha): {type_1_error}")
    print(f"Type 2 Error (beta): {type_2_error}")

# Example usage
population_mean = 50
population_std = 5
sample_size = 30
true_mean = 52
visualize_errors(population_mean, population_std, sample_size, true_mean)
```

###6.Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy import stats
# Sample data for two groups
group1 = [2.3, 3.1, 4.7, 5.6, 7.2, 8.4]
group2 = [4.2, 5.1, 6.7, 8.2, 9.5, 10.3]
# Perform independent T-test
t_statistic, p_value = stats.ttest_ind(group1, group2)
# Print the results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("We reject the null hypothesis. There is a significant difference between the two groups.")
else:
    print("We fail to reject the null hypothesis. There is no significant difference between the two groups.")
```

###9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean
    margin_of_error = sem * stats.t.ppf((1 + confidence) / 2, len(data) - 1)
    return mean - margin_of_error, mean + margin_of_error
# Example usage
sample_data = [2.3, 3.1, 4.7, 5.6, 7.2, 8.4]
ci = confidence_interval(sample_data)
print(f"Confidence interval: {ci}")
```

###11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
# Prior belief: Uniform distribution (assuming we know nothing about the bias)
def prior(theta):
    return 1 if 0 <= theta <= 1 else 0
# Likelihood function: Binomial distribution
def likelihood(theta, data):
    heads, tails = data
    return theta**heads * (1 - theta)**tails
# Posterior calculation using Bayes' Theorem
def posterior(theta, data):
    return likelihood(theta, data) * prior(theta)
# Normalize the posterior distribution
def normalize_posterior(posterior, theta_values):
    total_area = np.trapz(posterior, theta_values)  # Integrate to get the total area under the curve
    return posterior / total_area
# Generate theta values
theta_values = np.linspace(0, 1, 1000)
# Observed data: 9 heads and 3 tails
data = (9, 3)
# Calculate the posterior distribution
posterior_values = [posterior(theta, data) for theta in theta_values]
normalized_posterior_values = normalize_posterior(posterior_values, theta_values)
# Plot the prior, likelihood, and posterior distributions
plt.figure(figsize=(12, 6))
plt.plot(theta_values, [prior(theta) for theta in theta_values], label='Prior', linestyle='--')
plt.plot(theta_values, [likelihood(theta, data) for theta in theta_values], label='Likelihood', linestyle='-.')
plt.plot(theta_values, normalized_posterior_values, label='Posterior', linestyle='-')
plt.xlabel('Theta')
plt.ylabel('Probability Density')
plt.title('Bayesian Inference Using Bayes\' Theorem')
plt.legend()
plt.show()
```

###12.Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
# Create a contingency table
data = np.array([[50, 30], [20, 80]])
table = pd.DataFrame(data, columns=["Likes Dogs", "Does Not Like Dogs"], index=["Likes Cats", "Does Not Like Cats"])
# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(table)
# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)
# Interpret the results
alpha = 0.05
if p < alpha:
    print("We reject the null hypothesis. There is a significant association between the two variables.")
else:
    print("We fail to reject the null hypothesis. There is no significant association between the two variables.")
```

###16. Implement an F-test using Python to compare the variances of two random samples

```python
from scipy.stats import f
def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    f_statistic = var1 / var2
    dfn = len(sample1) - 1  # Degrees of freedom numerator
    dfd = len(sample2) - 1  # Degrees of freedom denominator
    p_value = 1 - f.cdf(f_statistic, dfn, dfd)
    return f_statistic, p_value
# Generate random samples
np.random.seed(0)
sample1 = np.random.normal(loc=10, scale=2, size=30)
sample2 = np.random.normal(loc=10, scale=4, size=30)
# Perform F-test
f_statistic, p_value = f_test(sample1, sample2)
# Print the results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("We reject the null hypothesis. The variances of the two samples are significantly different.")
else:
    print("We fail to reject the null hypothesis. The variances of the two samples are not significantly different.")
```

###19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
